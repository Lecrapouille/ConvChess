```python
%matplotlib inline
import numpy as np
import cPickle as pkl
import matplotlib.pyplot as plt
```

```python
losses = pkl.load(open("/data/convchess/src/net/models/regression/losses_g07_large.pkl"))
```

```python
def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]
```

```python
plt.plot(runningMeanFast(losses, 1000))
plt.show()
```
